# <center><h1 style="font-size:30px; color:Green;">"INTRODUCTION" </h1> </center>

<p style="font-family:Times New Roman (Headings CS);font-size:20px;line-height:1.5">The Apriori algorithm is given by R. Agrawal and R. Srikant in 1994 for finding frequent itemsets in a dataset for boolean association rules. The Name of the algorithm is Apriori because it uses prior knowledge of frequent itemset properties. </p>

<p style="font-family:Times New Roman (Headings CS);font-size:20px;line-height:1.5"> In this Algorithm we apply an iterative approach/level-wise search where k-frequent itemsets are used to find k+1 itemsets.</p>

<p style="font-family:Times New Roman (Headings CS);font-size:20px;line-height:1.5">To improve the efficiency of level-wise generation of frequent itemsets, an important property is used called Apriori property which helps by reducing the search space.</p>


# <center><h1 style="font-size:30px; color:MidnightBlue;">"Creating A Function using Apriory Hash Table Technique"</h1></center>



    This Function has 14 sub functions and 2 Main Functions which are used to find the Frequent Item Sets, Their rules and it will find the Support, Confidence and Lift as well. Our Main Functions are as under:

1. <b>My_apriory</b>.

2. <b>Generate_role_and_measure_support_confidence_lift</b>

<b>My_Apriory</b>

This function will take two parameters and return 2-D list of frequent items set using the approach of Hash Table Algorithm that is Apriory with Hash Table Algorithm to find Frequent Items set.

<b>All_Items_List:</b>

A 2-D List for all the transactions which will remove the NAN Values (Null values/empty values) from our dataset.

<b>Support:</b>

This is an integer parameter refers to minimum support value used to find Frequent Item Sets. 


<b>GENERATING RULES AND MEASURING SUPPORT AND CONFIDENCE</b>

		This Function take three parameters and save their output in a text file which we will give as a parameter and it will return all possible rules with their support, confidence and lift value. This functions have 3 parameters as under:
        
1. <b>Frequent_Items.</b>

2.	<b>All_Items_List.</b>

3. <b>file_name</b>

<b>Frequent Items:</b>
			
            This parameter takes 2-D list of Frequent Items Set returned by My_Apriory function.

<b>All_Items_Lis:</b>
			
            This is a 2-D list of all transactions which we load from our Dataset.
			After that we have applied all the functions on 3 different data sets respectively.

# <center><h3 style="font-size:25px; color:MidnightBlue;">"Our Self Made Functions for Apriori Algorithem using Hash Table"</h3></center>

```python
# 1
#call from 13
# return frequent item list
def pass_one(all_items_list, items_subset_from_last_pass, support):
    result = []
    itemset_size = 1
    if items_subset_from_last_pass: # check null or not
        itemset_size = len(items_subset_from_last_pass[0])
    item_counts = count_unique_items(all_items_list, items_subset_from_last_pass, support) # -> 3 return 2d list & count from 1st two parameters
    frequent_items = find_frequent_items(item_counts, support) # -> 4 return 2d list itemd grater then min "support" value
    if not frequent_items: # if frequent_items is empty
        return []
    result.append(frequent_items) # add frequent_items list in result list variable
    return result



# 2
#call from 13
# applay hashing technique
def pass_two(all_items_list, ans_one):
    result = []
    frequent_items = ans_one[0] # frequent items set from pas-one   
    result.append(frequent_items)    
    itemset_size = len(frequent_items[0]) + 1 # find the length of last items subset   
    pairs = generate_pairs(frequent_items, all_items_list, itemset_size + 1) # -> 5 pairs according pass-one items-subsets
    if not pairs: # if no more possiable pairs exist
        return result 
    bucket = maping_pairs_bucket(pairs, all_items_list) # -> 7 return buckets with total items count
    candidates = [] # store possiable pairs for next possiable subset
    for outer_list in bucket.values():
        for inner_list in outer_list:
            candidates.append(inner_list)
    result.append(bucket)
    result.append(candidates)
    return result



# 3
# single call by pass-one 
# return 2d list from original items set orr candiates items set
def count_unique_items(all_items_list, items_subset_from_last_pass, support):
    
    result = []    
    if not items_subset_from_last_pass: # in "items_subset_from_last_pass" is empty 
        items_and_count = {}  
        for single_transaction in all_items_list:
            for single_item in single_transaction: # use dictionary count douplicate values
                items_and_count.setdefault(single_item, 0)
                items_and_count[single_item] += 1

        #print("Count of All Items:\n", dict(sorted(ht_item.items())))
        for (item, count) in items_and_count.items(): # store each item and their count in a list
            result.append([item, count])
    else:
        result = items_subset_from_last_pass # return items from items_subset_from_last_pass (old items set)item set
    return result # return 2d list of orignal item count 



# 4
# call by pass-one 
# return frequent items set according to support
def find_frequent_items(item_counts, support):
    itemset_size = len(item_counts[0]) - 1
    frequent_items = []
    for item in item_counts:
        if item[itemset_size] >= support:
            frequent_items.append(item[:-1])
    frequent_items.sort()
    return frequent_items



# 5
#call in both
# get frequent items in 2d list and complete itemset in 2d list
# convert itemset into sub item set n-1,n-3,n-4 so....
def generate_pairs(frequent_items, all_items_list, itemset_size):
    result = []
    frequent_singletons = get_single_list(frequent_items) # -> 6 return single sorted list from 2d frequent_items list
    for frequent_item in frequent_items:
        fip = {}
        for single_transaction in all_items_list: # one by one get 1d item list from 2d items list 
            if len(single_transaction) < itemset_size + 1: # itemset_size use for subset(1st time it's 3(2+1), than 4 ...)
                continue
            inside = True
            for item in frequent_item: # it's 2d list pass inner list one by one
                if item not in single_transaction: # thst single frequent item not in that specific orignal item subset                    
                    inside = False
                    break
            if inside == True:
                for item in single_transaction: # get lingle item from 1d list of items
                    if (item not in frequent_singletons) or (item in frequent_item): # add in dictionary if this item in "frequent_singllist"
                        continue
                    fip.setdefault(item, 0)
        keys = fip.keys()
        keys = sorted(keys)
        for key in keys: # items name are in keys list variable
            tmp = list(frequent_item)
            tmp.extend(key)
            tmp.sort()
            if tmp not in result:
                result.append(tmp)
        
    true_result = []
    for items in result:
        inside = True
        for i in range(len(items)):
            tmp = list(items)
            tmp.pop(i)
            if tmp not in frequent_items:
                inside = False # if not a "frequent_items"
                break
        if inside == True: # check result items in "frequent_items" and add in one by one 
            true_result.append(items)
    #print("\n\nsubset of items which support is above from min-support: \n", sorted(true_result))
    return sorted(true_result)



# 6
#return 1d sorted unique items list 
def get_single_list(itemsets):
    temp = {}
    for items in itemsets:
        for item in items:
            temp.setdefault(item, 0) # all items become keays and its can not duplicate
    keys = temp.keys()
    keys = sorted(keys)
    return keys



# 7
# call from 11
# return a bucket dictionary and map each subset into bucket index
def maping_pairs_bucket(pairs, all_items_list):
    pair_size = len(pairs[0])
    item_and_no = assign_numbering(pairs) # -> 8 return all pair with a sequence numbering in dictionary
    count_pairs(pairs, all_items_list) # -> 9 count pair and store count value in orignal pairs variable
    bucket = {} # key: bucket_no, value: pairs
    for pair in pairs:
        bucket_no = 1
        for i in range(pair_size): # 
            c = pair[i]
            bucket_no *= item_and_no[c] # multiplay with assigning sequence number
        bucket_no %= 7 # calculate index number
        bucket.setdefault(bucket_no, []) # add into dictionary
        bucket[bucket_no].append(pair)
    return dict(sorted(bucket.items()))



# 8
# call from 7
# get 2d list and return dictionary and with number 1 to so on..
def assign_numbering(pairs):
    item_and_no = {}
    number = 1
    for pair in pairs:      # get pair set from 2d list one by one  =O(n), since length of pair is constant
        for c in pair:    # get single from 1d list
            if c not in item_and_no.keys():
                item_and_no.setdefault(c, number)
                number += 1
    #print("\n\nAssign sequence number: \n", dict(sorted(item_and_no.items())))
    return dict(sorted(item_and_no.items()))



# 9
# call by 7
# get pairs and all_items_list as 2d list 
# store count value in orinal pairs 2d list variable
def count_pairs(pairs, all_items_list):
    # count the pair in all_items_list and add at the last index
    itemset_size = len(pairs[0])
    for pair in pairs:
        count = 0
        for single_transaction in all_items_list:
            if len(single_transaction) < itemset_size:
                continue
            inside = True
            for i in range(len(pair)):
                if pair[i] not in single_transaction:
                    inside = False
                    break
            if inside == True:
                count += 1
        pair.append(count)
    #print("\n\ncount each subset and their frequency at the last index: \n",pairs)

        
# 10
# call from 14
# get 2d list and create 1d list with all unique frequent items
def frequent_items_in_onelist(frequent_items):
#     print("\n\nfrequent_items_in_onelist:")
#     print("\nfrequent_items:", frequent_items)
    result = {}
    if len(frequent_items) > 1:
#         print("\nif len(frequent_items) > 1:")
        for items in frequent_items:
#             print("\nfor items in frequent_items:", items)
            if len(items) >= 1:
#                 print("\nif len(items) >= 1:", items)
                for item in items:
                    result.setdefault(item, 0)
            #result.setdefault(item, 0)
        result = list(result.keys())
    else:
        result = frequent_items[0]
                  
    return result

# 11
# call from 14
# return 2d list of frequent items pair
def generate_frequent_items_pair(fitems):
    pairs = []
    list_len= len(fitems)
    for i in range(list_len):

        if i+1 < list_len:
            pairs.append([fitems[i], fitems[i+1]])
            #pairs.append([fitems[i], fitems[i+1], fitems[(list_len - 1) - i]])
        pairs.append([fitems[(list_len - 1) - i],fitems[i]])

    return pairs




# 12
# call from 14
def roules_and_support_confidence_lift(items_pair, all_items_list):
    
    rule = {'rule': [], 'support' : [], 'confidence' : [], 'lift' : []}
    tolaTransactions = len(all_items_list)
    
# 12.1  
# call from 12
# return the frequency or support
    def count_frequency(pairs, all_items_list):
    # count the pair in all_items_list and add at the last index
        itemset_size = len(pairs[0])
        for pair in pairs:
            count = 0
            for single_transaction in all_items_list:
                if len(single_transaction) < itemset_size:
                    continue
                inside = True
                for i in range(len(pair)):
                    if pair[i] not in single_transaction:
                        inside = False
                        break
                if inside == True:
                    count += 1
            return count
# 12.2    
# call from 12
# add new rule and support_confidence_lift in dictionary
    def support_confidence_lift(a, b , ab):
        
        frequency_A = count_frequency([a], all_items_list) # -> 12.1 get frequency
        frequency_B = count_frequency([b], all_items_list) # -> 12.1 get frequency
        frequency_AB = count_frequency([ab], all_items_list) # -> 12.1 get frequency
        if frequency_AB < 2:
            return 0

        support_A = frequency_A / tolaTransactions # -> 12.1 get frequency
        support_B = frequency_B / tolaTransactions # -> 12.1 get frequency
        support_AB = frequency_AB / tolaTransactions
        
        try:
            r = str(a)[1:-1] + " -> " + str(b)[1:-1]
            s = support_AB
            c = frequency_AB / frequency_A               
            l = c / support_B 
        except ZeroDivisionError:
            return 0
        rule['rule'].append(r) 
        rule['support'].append(s)
        rule['confidence'].append(c)
        rule['lift'].append(l)


    for pair in items_pair:
        if len(pair) == 1:
            continue

        if len(pair) == 2:
            a = pair[:1]
            b = pair[1:]
            ab = pair
            support_confidence_lift(a, b , ab) # -> 12.2 get support_confidence_lift
#         if len(pair) == 3:  
#             length = (len(pair))
#             a = pair[:length]
#             b = pair[:length-1]
#             ab = pair 
#             support_confidence_lift(a, b , ab) # -> 12.2 get support_confidence_lift

    return rule



# 13
def my_apriory(all_items_list, support):

    frequent_items = []
    items_subset_from_last_pass = [] # store subset and their count

    # remove "NaN" from items list
    all_items_list = [[x for x in y if str(x) != 'nan'] for y in all_items_list]
    while 1:
        ans_one = pass_one(all_items_list, items_subset_from_last_pass, support)
        if not ans_one: # if not frequent item found
            break
        ans_two = pass_two(all_items_list, ans_one)
        frequent_items = ans_two[0]
        if len(ans_two) == 1: # if ans-two have not any subset for next frequent pair 
            break
        items_subset_from_last_pass = ans_two[2]
    return frequent_items



# 14
def generate_role_and_measure_support_confidence_lift(frequent_items, all_items_list, file_name):
    # remove "NaN" from items list
    all_items_list = [[x for x in y if str(x) != 'nan'] for y in all_items_list]

    f = frequent_items_in_onelist(frequent_items) # -> 10 send 2d list and get 1d unique items list
    f_pairs = generate_frequent_items_pair(f) # -> 11 return 2d list of frequent_items in pairs
    rule = roules_and_support_confidence_lift(f_pairs, all_items_list) # -> 13 get all rule and their support_confidence_lift
    
    # store all data into text file
    with open(file_name, 'w') as file:
        text = "rule, support, confidence, lift\n"
        for i in range(len(rule["rule"])):
            text += '[' + str(rule["rule"][i]) +']' + ', ' + str(rule["support"][i]) + ', ' + str(rule["confidence"][i]) + ', ' + str(rule["lift"][i]) + '\n'
        file.write(text + '\n')
        print("Result also save in text file successfully..!!")
        
    # pd.DataFrame(rule).reindex(pd.DataFrame(rule).rule.str.len().sort_values().index).reset_index(drop=True)
    df = pd.DataFrame(rule) # dictionary convert info dataframe 
    try:
        s = df.rule.str.len().sort_values().index # sort as rules base it's gove index number list  
        df = df.reindex(s) # sort according list
        df = df.reset_index(drop=True) # index number in range
    except:
        return df
    return df

    
```
